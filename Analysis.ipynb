```python
# load all_the_features_combined_df.pkl

import pandas as pd

all_the_features_combined_df = pd.read_pickle('all_the_features_combined_df.pkl')
```

```python
all_the_features_combined_df.head(5)
```

```python
# load all_fouls.pkl and testing_fouls.pkl

all_fouls = pd.read_pickle('all_fouls.pkl')

testing_fouls = pd.read_pickle('testing_fouls.pkl')
```

```python
# combine all_fouls and testing_fouls into fouls

fouls = pd.concat([all_fouls, testing_fouls])
```

```python
fouls.head(15)
```

```python
# set 'id' as index for fouls

fouls.set_index('id', inplace=True)
```

```python
# in all_the_features_combined_df, create a columns called 'match_id', 'possession_team_id', 'possesion_team', 'player_id', 'player'. Fill it from the fouls dataframe. Both dataframes have 'id' as its index.

all_the_features_combined_df['match_id'] = fouls['match_id']

all_the_features_combined_df['possession_team_id'] = fouls['possession_team_id']

all_the_features_combined_df['possession_team'] = fouls['possession_team']

all_the_features_combined_df['player_id'] = fouls['player_id']

all_the_features_combined_df['player'] = fouls['player']
```

```python
# remove columns: foul_committed_advantage	foul_committed_offensive	foul_committed_penalty	foul_committed_card	foul_committed_counterpress	foul_committed_type	foul_won_advantage	foul_won_defensive	foul_won_penalty from fouls dataframe and save it as fouls_short

fouls_short = fouls.drop(['foul_committed_advantage', 'foul_committed_offensive', 'foul_committed_penalty', 'foul_committed_card', 'foul_committed_counterpress', 'foul_committed_type', 'foul_won_advantage', 'foul_won_defensive', 'foul_won_penalty'], axis=1)

fouls_short.head(10)
```

```python
from statsbombpy import sb

comp = sb.competitions()
comp = comp[comp.season_name != '1999/2000']

all_matches = pd.DataFrame()

for index, row in comp.iterrows():
    matches = sb.matches(row['competition_id'], row['season_id'])
    all_matches = all_matches.append(matches)
```
